## IA mode

Nous avons trois IA dans ce projet:
- l'ia stupide
- l'ia heuristique
- l'ia smart

l'ia stupide choisi alatoirement les cases  jouer sans tenir compte d'avoir touch un bateau ou non

l'ia heuristique choisi alatoirement les cases jusqu' toucher un bateau, l'ia passe alors en mode track et choisi les cases adjacantes, lorsque le bateau est coul, s'il n'y a plus de point  couler elle recommence  cherhcer alatoirement.

l'ia Smart marche comme l'ia heuristique sauf que la phase de recherche alatoire ne prend que des points distants de 2 entre eux, comme si la grille tait un chiquier et l'IA ne cible que les cases noirs.

```python
N_PARTIES = 1000
for ia_couple in [(player.DUMB, player.HUNTER),(player.DUMB, player.ULTIMATE),((player.HUNTER, player.ULTIMATE))]:
    score =  sum(battle_ship.play_versus_nb_party(ia_couple[0], ia_couple[1],  N_PARTIES))
    score_in_percent = score / N_PARTIES
    print(f"{score} parties gagnes de {ia_couple[0]} sur {ia_couple[1]} pour un total de {N_PARTIES} - {'{:.2f}'.format(score_in_percent)}")
```

#### nombre de coup pour gagner

```python
def create_data_battle_ship():
    battle_ship = BattleShip()
    nb_index = 10000
    liste_index = [i for i in range(1, nb_index)]
    liste_nb_coup_ia_Dumb = [battle_ship.play_one_game(player.DUMB)[1] for _ in liste_index]
    liste_nb_coup_ia_Hunter = [battle_ship.play_one_game(player.HUNTER)[1] for _ in liste_index]
    liste_nb_coup_ia_Ultime = [battle_ship.play_one_game(player.ULTIMATE)[1] for _ in liste_index]
    
    
    return pandas.DataFrame({'nb_party': liste_index , 'nb_coup_ia_dumb': liste_nb_coup_ia_Dumb, 
                             'nb_coup_ia_Hunter': liste_nb_coup_ia_Hunter,
                            'nb_coup_ia_ultime': liste_nb_coup_ia_Ultime}, 
                          columns = ['nb_party', 'nb_coup_ia_dumb', 'nb_coup_ia_Hunter', 'nb_coup_ia_ultime'])
ddb = create_data_battle_ship()
ddb.describe()
```

l'analyse des donnes montre qu'en moyenne l'ia dumb russi une partie avec 95 coups au minimum sur 100.
ce qui n'est pas trs tonnant.


L'Ia hunter et ultime ont des moyennes assez proche l'un de l'autre. 59 pour l'hunter et 51 pour l'ultime.
La proximit de ces 2 moyennes ne permet pas  l'Ia ultime de gagn quasiment tout le temps fasse  l'ia Hunter, par contre la diffrence entre l'ia dumb et ultime assure une quasi-victoire pour l'ia Ultime  


L'ia ultime russi dans le pire des cas en 70 coups, l ou les 2 autres Ia ont un score de 100. la tactique de l'ia ultime pour recherche les cases est donc beaucoup plus efficace dans le pire des cas. 

#### nombre de victoire par IA

```python
bt = BattleShip()
nb_index = 500
liste_index = [i for i in range(1, nb_index)]
tps1 = time.time()
vic_D_H = [sum(bt.play_versus_nb_party(player.DUMB, player.HUNTER, i))/i for i in liste_index]
tps2 = time.time()
print(f"temps combat Ia Dumb - Ia Hunter {tps2 - tps1}")
tps1 = time.time()
vic_D_S = [sum(bt.play_versus_nb_party(player.DUMB, player.ULTIMATE, i))/i for i in liste_index]
tps2 = time.time()
print(f"temps combat Ia Dumb - Ia Smart {tps2 - tps1}")
tps1 = time.time()
vic_H_S = [sum(bt.play_versus_nb_party(player.HUNTER, player.ULTIMATE, i))/i for i in liste_index]
tps2 = time.time()
print(f"temps combat Ia Hunter - Ia Smart {tps2 - tps1}")
vic_S_D = [1 - vic_D_S[i] for i in range(nb_index - 1)]
vic_H_D = [1 - vic_D_H[i] for i in range(nb_index - 1)]
vic_S_H = [1 - vic_H_S[i] for i in range(nb_index - 1)]
    
df_combat =  pandas.DataFrame({'nb_party': liste_index , 
                               'vic_D_H': vic_D_H, 
                             'vic_D_S': vic_D_S,
                            'vic_H_S': vic_H_S,
                              'vic_S_D':vic_S_D,
                              'vic_H_D':vic_H_D,
                              'vic_S_H':vic_S_H},
                          columns = ['nb_party', 'vic_D_H', 
                                     'vic_D_S', 'vic_H_S',
                                    'vic_S_D', 'vic_H_D',
                                    'vic_S_H'])
df_combat.describe()
```

### graphiques

#### nombre de coup pour gagner

```python
def create_data_battle_ship_nb_coup():
    tps1 = time.time()
    battle_ship = BattleShip()
    NB_PARTIE = 500
    liste_index = [i for i in range(1, NB_PARTIE)]
    liste_nb_coup_ia_Dumb = [battle_ship.play_nb_game(player.DUMB, i) for i in liste_index]
    tps2 = time.time()
    print(f"temps pour calculer le nombre de coup pour Ia Dumb: {tps2 - tps1}")
    tps1 = time.time()
    liste_nb_coup_ia_Hunter = [battle_ship.play_nb_game(player.HUNTER, i) for i in liste_index]
    tps2 = time.time()
    print(f"temps pour calculer le nombre de coup pour Ia Hunter: {tps2 - tps1}")
    tps1 = time.time()
    liste_nb_coup_ia_Ultime = [battle_ship.play_nb_game(player.ULTIMATE, i) for i in liste_index]
    tps2 = time.time()
    print(f"temps pour calculer le nombre de coup pour Ia Smart: {tps2 - tps1}")
    tps1 = time.time()
    return pandas.DataFrame({'nb_party': liste_index , 'nb_coup_ia_dumb': liste_nb_coup_ia_Dumb, 
                             'nb_coup_ia_non_Hunter': liste_nb_coup_ia_Hunter,
                            'nb_coup_ia_ultime': liste_nb_coup_ia_Ultime}, 
                          columns = ['nb_party', 'nb_coup_ia_dumb', 'nb_coup_ia_non_Hunter', 'nb_coup_ia_ultime'])
df = create_data_battle_ship_nb_coup()

df['nb_coup_ia_dumb'] = df['nb_coup_ia_dumb'].div(df['nb_party'], axis = 0)
df['nb_coup_ia_non_Hunter'] = df['nb_coup_ia_non_Hunter'].div(df['nb_party'], axis = 0)
df['nb_coup_ia_ultime'] = df['nb_coup_ia_ultime'].div(df['nb_party'], axis = 0)

fig1, axs1 = plt.subplots(nrows=1, ncols=1, figsize=[10,5])
axs1.set_xlabel('nombre de partie', fontsize=15)
axs1.set_ylabel('moyenne du nombre de coup jou', fontsize=15)
axs1.set_title('moyenne par IA /nombre de partie', fontsize=15)
df['nb_coup_ia_dumb'].plot(x=df['nb_party'], label='iaDumb')
df['nb_coup_ia_non_Hunter'].plot(x=df['nb_party'], label='iaHunter')
df['nb_coup_ia_ultime'].plot(x=df['nb_party'], label='IaSmart')
axs1.legend()
```
